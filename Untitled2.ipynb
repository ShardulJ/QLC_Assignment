```python
text = df['ActivityDescription'].values.tolist()
preprocess_text = []
```

```python
def clean_text(text):
  text = re.sub('[!\"#$%&\'()*+,-./:;<=>?@[\]^_`{|}~0-9\n]', " ", str(text)).lower()
  text_tokens = word_tokenize(text)
  tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
  return tokens_without_sw
```

```python
for txt in tqdm(range(len(text))):
  preprocess_text.append(clean_text(text[txt]))
```

```python
df['msg'] = df['ActivityDescription'].apply(lambda x : re.sub('[^a-zA-Z0-9\s]',"",str(x)))
```

```python
def produce_wordcloud(rest):
    
    plt.figure(figsize=(20,30))
    for i,r in enumerate(rest):
        plt.subplot(3,3,i+1)
        corpus=df[df['OpportunityStageName']==r]['msg'].values.tolist()
        corpus=' '.join(x  for x in corpus)
        wordcloud = WordCloud(max_font_size=None, background_color='white', collocations=False,
                      width=1500, height=1500).generate(corpus)
        plt.imshow(wordcloud)
        plt.title(r)
        plt.axis("off")
        

        
rest=df['OpportunityStageName'].value_counts()[:9].index        
produce_wordcloud(rest)
```
